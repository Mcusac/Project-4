```python
import pandas as pd
```

```python
df = pd.read_csv('dataset1.csv')
df.head(20)
```

```python
df = df.query('mileage != 0')
df.head(20)
```

```python
df.drop_duplicates()
df
```

```python
df = df.dropna()
df
```

```python
df['engine string'] = df['engine']
```

```python
for i in range(7604):
    string = df.iloc[i, 6]
    string = string.replace('-', ' ')
    string = string.replace('cc', ' cc')
    string = string.replace('ci', ' ci')
    string = string.replace('L', ' L')
    string = string.replace(',', '')
    string = string.replace('.', '.')
    df.iloc[i,6] = string
df
```

```python
for i in range(7604):
    string = df.iloc[i, 6]
    try:
        aList = string.split()
        li = 0
        for j in aList:
            try:
                j = float(j)
            except:
                pass
            if type(j) == float:
                try:
                    if aList[li + 1].lower() == 'liter':
                        df.iloc[i, 6] = j * 1000
                    elif aList[li + 1].lower() == 'l':
                        df.iloc[i,6] = j * 1000
                    elif aList[li + 1].lower() == 'cc':
                        df.iloc[i,6] = j
                    elif aList[li + 1].lower() == 'ci':
                        df.iloc[i,6] = j * 16.387
                except:
                    df.iloc[i,6] = 0
            li = li + 1
    except:
        pass
```

```python
for i in range(7604):
    value = df.iloc[i, 6]
    if type(value) == str:
        df.iloc[i, 6] = 0
```

```python
df['engine'] = df['engine'].astype(float)
```

```python
df = df.query('engine != 0')
df
```

```python
df = df.query('mileage < 300000')
df
```
